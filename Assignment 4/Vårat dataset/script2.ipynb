```python
from matplotlib import pyplot as plt
import pandas as pd
from datetime import datetime, timedelta

df = pd.read_json('/Users/ludvigflodin/Documents/Courses/D7018B/Assignment 4/Vårat dataset/LudvigAlfredDataset.json', lines=True)
print(df.shape)
df.head(-1)
# df = df.drop(df.index[::2]).reset_index(drop=True)

```

```python
fig, axs = plt.subplots(4, 1, figsize=(12, 12), sharex=True)
datetime = 'Date and time (GMT+1)'
# Temperature
axs[0].plot(df[datetime], df["Temperature (°C)"], label="Temperature (°C)", color="tab:red")
axs[0].set_ylabel("Temperature (°C)")
# axs[0].legend(loc="upper right")
axs[0].grid(True)

# Moisture
axs[1].plot(df[datetime], df["Moisture (0-1)"], label="Moisture (%)", color="tab:blue")
axs[1].set_ylabel("Moisture (%)")
# axs[1].legend(loc="upper right")
axs[1].grid(True)

#Humidity
axs[2].plot(df[datetime], df["Humidity (%)"], label="Humidity (%)", color="tab:green")
axs[2].set_ylabel("Humidity (%)")
# axs[2].legend(loc="upper right")
axs[2].grid(True)

# Luminance
axs[3].plot(df[datetime], df["Luminance (0-1)"], label="Lumninace (%)", color="tab:orange")
axs[3].set_ylabel("Luminance (%)")
axs[3].set_xlabel("Time")
# axs[3].legend(loc="upper right")
axs[3].grid(True)

fig.suptitle("Sensor data over time", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
